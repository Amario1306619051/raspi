Metode yang dapat digunakan antara lain :

Gradient threshold dapat digabung dengan color threshold dan mungkin saja menghasilkan penglihatan yang lebih baik untuk mendeteksi marka jalan 

<hr>
<h1> Canny edges </h1>


```python
img = cv2.imread('lane-detection-using-machine-learning.webp')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('ruang warna grayscale')
```

```python
def canny_detector(image, low_threshold, high_threshold):
    """
    Apply Canny Edge Detection algorithm to the input image.
        Parameters:
            image: An np.array compatible with plt.imshow.
            low_threshold (Default = 50).
            high_threshold (Default = 150).
    """
    return cv2.Canny(image, low_threshold, high_threshold)

gray1 = canny_detector(gray, 50, 150)
```

```python
plt.imshow(gray1, cmap='gray')
plt.title('lt = 50 ht=150')
```

<h2> Mengubah parameter threshold </h2>

Langkah ini dilakukan dengan diharapkan hanya lajur yang terdeteksi



```python
gray2 = canny_detector(gray, 185, 500)
plt.imshow(gray2, cmap='gray')
plt.title('lt = 185 ht= 500')
```

```python
gray2 = canny_detector(gray, 250, 400)
plt.imshow(gray2, cmap='gray')
plt.title('lt = 250 ht= 400')
```

Berdasarkan dua gambar diatas, gambar dengan low threshold 250 dan high threshold = 400 lebih baik dalam mendeteksi marka jalan
<hr>

<h1> Sobel edge detection </h1>

```python
grayblur = cv2.GaussianBlur(gray,(3,3),0)
grayblur = cv2.Sobel(grayblur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)
```

```python
plt.imshow(grayblur, cmap='gray')
plt.title('lt = 250 ht= 400')
```

```python
graySobel = cv2.Sobel(gray, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)
plt.imshow(graySobel, cmap='gray')
plt.title('lt = 250 ht= 400')
```
